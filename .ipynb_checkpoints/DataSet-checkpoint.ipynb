```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
import pandas as pd

# Cargar el dataset
data = pd.read_csv('/Users/cristiangaytan/Downloads/COVID-19_Daily_Counts_of_Cases__Hospitalizations__and_Deaths_20241016.csv')

# Verificar las columnas disponibles
print(data.columns)

# Seleccionar dos columnas para graficar (ajusta según tu dataset)
x = data['CASE_COUNT']  # Reemplaza 'Column1' con el nombre real
y = data['date_of_interest']  # Reemplaza 'Column2' con el nombre real

# Crear la gráfica de dispersión
plt.scatter(x, y)
plt.xlabel('date_of_interest')  # Etiqueta para el eje X
plt.ylabel('CASE_COUNT')  # Etiqueta para el eje Y
plt.title('Gráfica de dispersión del dataset')
plt.show()
```
